```
#Question 2
#Download the dataset from this link :
#https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv
 
#Import the data and print the following :
#a.) First 5 rows of Dataframe 
#b.) First 10 rows of the Dataframe 
#c.) find basic statistics on the particular dataset. 
#d.) Find the last 5 rows of the dataframe 
#e.) Extract the 2nd column and find basic statistics on it.


import pandas as pd

#Reading CSV file from the given link
df=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv")

#Printing first five rows of dataframe(no argument inside head() automatically prsents first 5 rows)
firstFiveRows=df.head()
print("\nFirst Five Rows are:\n\n",firstFiveRows)

print("\n\n------------------------------------------------------------------\n\n")

#Printing first ten rows of dataframe(head(n) gives first n rows of dataframe)
firstTenRows=df.head(10)
print("\n\nFirst Ten Rows are:\n\n",firstTenRows)

print("\n\n------------------------------------------------------------------\n\n")

#Basic statistics of dataframe(describe()gives complete detail of the dataset)
statistics=df.describe(include="all")
print("\n\nStatistics of the dataset:\n\n",statistics)

print("\n\n------------------------------------------------------------------\n\n")

#Last five rows of dataframe
lastFiveRows=df.tail()
print("\n\nLast Five Rows are:\n\n",lastFiveRows)

print("\n\n------------------------------------------------------------------\n\n")

#Statistics of Second column
secColStats=df['Location'].describe(include='all')
print("\n\nSecond Column Statistics:\n\n",secColStats)

```
