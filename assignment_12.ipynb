```
import matplotlib.pyplot as plt
import sklearn.metrics as mt
%matplotlib inline
```

```
from sklearn.datasets import load_digits
digits = load_digits()
```

```
for i in range(0,10):
  plt.imshow(digits.images[i], cmap=plt.cm.gray_r, interpolation='nearest')
  plt.show()
```

```
X=digits.data
y=digits.target
```

```
# Print to show there are 1797 images (8 by 8 images for a dimensionality of 64)
print("X shape", X.shape)

# Print to show there are 1797 labels (integers from 0-9)
print("Label shape", y.shape)
```

```
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```
x_train.shape
```

```
x_test.shape
```

```
from sklearn.linear_model import LogisticRegression
```

```
reg = LogisticRegression()
```

```
reg.fit(x_train, y_train)
```

```
predict_op=reg.predict(x_test)
print(predict_op)
```

```
# Use score method to get accuracy of model
score = reg.score(x_test, y_test)
print(score)
```

## Classification accuracy:

```
class_accuracy=mt.accuracy_score(y_test, predict_op)
print(class_accuracy)
```

## Logarithmic loss:

```
# We are using the prediction and not the probability of your prediction. 
# The predict_op variable contains [ 1 2 1 3 .... ] but to use log_loss it should contain 
# something like [[ 0.1, 0.8, 0.1] [ 0.0, 0.79 , 0.21] .... ]. to obtain these 
# probabilities use the function predict_proba:

predict_prob =reg.predict_proba(x_test)
lloss = mt.log_loss(y_test,predict_prob) 
print(lloss)
```

## R2 metrics:

```
r2=mt.r2_score(y_test,predict_op)
print(r2)
```

## Mean Absolute Error:

```
from sklearn.metrics import mean_absolute_error
```

```
mae=mean_absolute_error(y_test, predict_op)
print(mae)
```

## Mean Squared Error:

```
from sklearn.metrics import mean_squared_error
```

```
mse=mean_squared_error(y_test, predict_op)
print(mse)
```

## Root Mean Squared Error:

```
import math
```

```
rmse=math.sqrt(mse)
print(rmse)
```

## Confusion Matrix:

```
from sklearn.metrics import confusion_matrix
```

```
con_matrix=confusion_matrix(y_test,predict_op)
print(con_matrix)
```

## Classification Report:

```
class_report=mt.classification_report(y_test,predict_op)
print(class_report)
```
