```
#Question_1
#mean of elements

import numpy as np

narray=np.random.rand(10,1)
print("Array elements are\n",narray)

mean_narray=np.mean(narray)
print("\nMean:",mean_narray)
```

```
#Question_2
#Variance and Standard Deviation

import numpy as np

narray=np.random.rand(20,1)
print("Array elements are\n",narray)

variance_narray=np.var(narray)
print("\nVariance is:",variance_narray)

sd_narray=np.std(narray)
print("Standard Deviation:",sd_narray)
```

```
#Question 3
#matrx multiplication and summation of elements of new matrix formed

narrayA=np.random.rand(10,20)
print("Array A elements are\n\n",narrayA)
print("Dimension of matrix A:", np.shape(narrayA))

narrayB=np.random.rand(20,25)
print("\nArray B elements are\n\n",narrayB)
print("Dimension of matrix B:", np.shape(narrayB))

#multiplication of matrices using np.dot()
print("\n\n\n Matrix Multiplication\n")
multiplication=np.dot(narrayA,narrayB)
print(multiplication)      
print("Dimension of multiplied matrix is",np.shape(multiplication))

#addition of matrices
summation=np.sum(multiplication)
print("\n\nSum of all the elements of new matrix:", summation)
```

```
#Question 4
#Applying exp() in the array elements

narray=np.random.rand(10,1)
narrayExp=1/(1+np.exp(-narray))
print("Original array:\n\n",narray)
print("\n\nAfter applying exponential function\n\n",narrayExp)
```
