```
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns

```

```
#importing dataset
df=pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv')
df.head()
```

```
plot=sns.countplot(x="Sex",data=df)
#setting title for the graph
plot.set_title("Count of Male and Female")
sns.despine()
```

```
plot=sns.countplot(x="Pclass",data=df)
#setting title for the graph
plot.set_title("Count of different class people")
sns.despine()
```

```

plot=sns.countplot(x="Age",data=df)
#setting title for the graph
plot.set_title("Count of Age of people")
sns.despine()
```

```
plot=sns.boxplot(x="Sex",y="Age",data=df)
#setting title for the graph
plot.set_title("Comparison of Sex with Age")
sns.despine()
```

```
plot=sns.violinplot(x="Survived",y="Age",data=df)
#setting title for the graph
plot.set_title("Comparison of Age and Survived")
sns.despine()
```

```
plot=sns.jointplot(x="Survived",y="Pclass",data=df)
plt.title("comparison of survived and Pclass")
sns.despine()
```

```
countplt=sns.jointplot(x="Age",y="Survived",data=df)
# set_title does not work with joint plot
#setting title for the graph
plt.title("Comparison of Age and Survived")
sns.despine()
```

```
countplt=sns.jointplot(y="Fare",x="Survived",data=df)
# set_title does not work with joint plot
#setting title for the graph
plt.title("Comparison of Fare and Survived")
sns.despine()
```

```
plot=sns.boxplot(x="Survived",y="Age",data=df)
#setting title for the graph
plot.set_title("Comparison of Survived with Age")
sns.despine()
```

```
plot=sns.violinplot(x="Survived",y="Pclass",data=df)
#setting title for the graph
plot.set_title("Comparison of Pclass and Survived")
sns.despine()
```

```
plot=sns.distplot(df['Fare'])
# set_title does not work with joint plot
#setting title for the graph
plot.set_title("Plotting of Fare")
sns.despine()
```

```
plot=sns.distplot(df['PassengerId'])
plot.set_title("Plotting Passenger ID ")
sns.despine()
```

```
plot=sns.barplot(x="Survived",y="Age",data=df)
#setting title for the graph
plot.set_title("Comparison of Survived and Age")
sns.despine()
```
